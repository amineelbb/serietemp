Qu’est ce que la stationnarité? 

La stationnarité dans les séries temporelles signifie que les données n'ont pas de tendance ou de schéma qui change avec le temps. Cela facilite l'analyse et la prédiction des données temporelles.

Visuellement, pour vous la time serie est-elle stationnaire ou non stationnaire? 


visuellement la time serie semble stationnaire car elle semble fluctuer autour de la moyenne mobile 

Réalisez un test de Dickey Fuller, quelle est l’hypothèse nulle et l’hypothèse alternative, quelle est la statistique calculée et quelle distribution théorique suit elle d’après l’hypothèse nulle. Quelle est la probabilité que la statistique sous l’hypothèse H0 prenne un valeur aussi extrême voir plus extrême? 

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller

# Chargez votre série temporelle (remplacez ceci par vos données)
# Exemple de données : une liste de valeurs
print("L'hypothèse nulle HO est le fait que la time serie n'est pas stationnaire ")

print("L'hypothèse alternative H1 est le fait que la time serie est stationnaire ")

serie_temporelle = data_by_day.values
# Réalisez le test de Dickey-Fuller
resultat_test = adfuller(serie_temporelle)

# Récupérez les résultats du test
statistique_test = resultat_test[0]
p_valeur = resultat_test[1]
valeurs_critiques = resultat_test[4]

# Affichez les résultats
print(" La statistique calculée dans le test de Dickey-Fuller est appelée la statistique t de Dickey-Fulle,Plus la statistique t est négative et plus sa valeur absolue est élevée, plus la série temporelle est susceptible d'être stationnaire.")
print("Sous l'hypothèse nulle (H0) de non-stationnarité, la statistique t de Dickey-Fuller suit une distribution de Student avec un certain nombre de degrés de liberté")
print("Statistique du test de Dickey-Fuller :", statistique_test)
print("Valeur p (p-valeur) :", p_valeur)
print("Valeurs critiques (niveau de confiance de 5%) :", valeurs_critiques)

# Interprétez les résultats
if p_valeur <= 0.05:
    print("Conclusion :comme la p_value est inferieur a 0.05, La série temporelle est stationnaire (rejet de l'hypothèse nulle).")
else:
    print("Conclusion : La série temporelle n'est pas stationnaire (non-rejet de l'hypothèse nulle).")
```

Même consigne pour le test de KPSS


```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import kpss

# Chargez votre série temporelle (remplacez ceci par vos données)
# Exemple de données : une liste de valeurs

print("L'hypothèse nulle HO est le fait que la time serie est stationnaire ")

serie_temporelle = data_by_day.values

# Réalisez le test de KPSS
resultat_test = kpss(serie_temporelle)

# Récupérez les résultats du test
statistique_test = resultat_test[0]
p_valeur = resultat_test[1]
lags_optimal = resultat_test[2]

# Affichez les résultats
print(" La statistique calculée dans le test de KPSS mesure la déviation de la série temporelle par rapport à la stationnarité. Plus la statistique est élevée, plus la série temporelle est susceptible d'être non stationnaire.")
print("Sous l'hypothèse nulle (H0) de non-stationnarité, la statistique du test de KPSS suit une distribution asymptotique, qui dépend de l'estimation de la variance et du nombre de lags utilisés dans le test")

print("Statistique du test de KPSS :", statistique_test)
print("Valeur p (p-valeur) :", p_valeur)
print("Lags optimal :", lags_optimal)

# Interprétez les résultats
if p_valeur <= 0.05:
    print("Conclusion : La série temporelle n'est pas stationnaire (rejet de l'hypothèse nulle).")
else:
    print("Conclusion : La série temporelle est stationnaire (non-rejet de l'hypothèse nulle).")
```

Désaisonnalisez la composante annuelle de votre time série est utilisez à nouveau le test de Dickey Fuller. Que comprenez vous de ce résultat?

```python
moyenne_mobile = data_by_day.rolling(window=12, center=True).mean()

# Soustraire la moyenne mobile de la série pour obtenir la série désaisonnalisée
serie_desaisonnalisee = data_by_day - moyenne_mobile

# Supprimer les valeurs NaN résultant de la moyenne mobile centrée
serie_desaisonnalisee = serie_desaisonnalisee.dropna()

# Réaliser le test de Dickey-Fuller sur la série désaisonnalisée
resultat_test = adfuller(serie_desaisonnalisee)

# Récupérer les résultats du test
statistique_test = resultat_test[0]
p_valeur = resultat_test[1]
valeurs_critiques = resultat_test[4]

# Afficher les résultats du test
print("Statistique du test de Dickey-Fuller pour la série désaisonnalisée :", statistique_test)
print("Valeur p (p-valeur) :", p_valeur)
print("Valeurs critiques (niveau de confiance de 5%) :", valeurs_critiques)

# Interpréter les résultats
if p_valeur <= 0.05:
    print("Conclusion : La série désaisonnalisée est dépendante de sa composante annuelle qui est responsable de sa stationnalité.la serie est donc stationnaire (rejet de l'hypothèse nulle).")
else:
    print("Conclusion : La série désaisonnalisée n'est pas stationnaire (non-rejet de l'hypothèse nulle).")
```

Qu’est ce que la métrique MAPE? 


La metrique  MAPE exprime l'erreur en pourcentage par rapport aux valeurs réelles. Plus le MAPE est faible, plus le modèle de prévision est considéré comme précis. 

```python
erreurs_absolues = np.abs(real_values_2021_2022 - forecast)
mape = np.mean(erreurs_absolues / real_values_2021_2022) * 100

print(f"Le MAPE est de {mape:.2f}%")
```

Quel est la particularité des train test split quand on travaille avec des séries temporelles?

Lorsque l'on travaille avec des séries temporelles, la division train-test doit respecter l'ordre chronologique des données pour simuler des prédictions futures. Il faut éviter la fuite de données en ne mélangeant pas accidentellement des informations futures dans les données d'entraînement.

Comment peut-on faire une cross validation avec des séries temporelles?


La cross-validation pour les séries temporelles implique de diviser les données chronologiquement, formant une fenêtre de validation mobile pour évaluer le modèle sur différentes périodes.

```python
import numpy as np
from darts import TimeSeries
from darts.models import NaiveDrift
from darts.metrics import mape

# Convertissez votre série désaisonnalisée en une série temporelle Darts
# Supposons que votre série désaisonnalisée soit dans une variable appelée 'serie_desaisonnalisee'

# Créez une série temporelle à partir de votre série désaisonnalisée
# serie_temporelle = TimeSeries.from_dataframe(serie_desaisonnalisee)

# Divisez la série en train et test (garder les deux dernières années en test)
train, test = serie_desaisonnalis.split_after(pd.Timestamp('2022-01-01'))  # Remplacez 'YYYY-MM-DD' par la date de début de vos données de test

# Créez un modèle de Naive Drift
model = NaiveDrift()

# Adaptez le modèle aux données d'entraînement
model.fit(train)

# Effectuez des prévisions sur les données de test
forecast = model.predict(len(test))

# Calculez la MAPE
mape_score = mape(test, forecast)

# Affichez la MAPE
print(f"MAPE: {mape_score}")

# Tracez la série temporelle et les prévisions
plt.figure(figsize=(12, 6))
plt.plot(serie_desaisonnalis, label='Séerie desaisonalisse', color='blue')
plt.plot(test, label='Données de test', color='green')
plt.plot(forecast, label='Prévisions', color='red')
plt.title('Série desaisonalise et prévisions avec Naive Drift')
plt.xlabel('Date')
plt.ylabel('Valeurs')
plt.legend()
plt.grid(True)
plt.show()
```
